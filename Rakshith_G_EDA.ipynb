## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize distribution of transaction values
plt.figure(figsize=(10, 6))
sns.histplot(transactions['TotalValue'], bins=30, kde=True)
plt.title('Distribution of Transaction Values')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.show()

# Visualize total sales by region
sales_by_region = transactions.groupby('CustomerID')['TotalValue'].sum().reset_index()
sales_by_region = sales_by_region.merge(customers[['CustomerID', 'Region']], on='CustomerID')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='TotalValue', data=sales_by_region)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Value')
plt.show()
```
